```python
import pandas as pd
```

```python
demo = pd.read_csv("../data/raw/demographic.csv")
```

```python
df_demo = demo[["RIDRETH3", "RIAGENDR", "RIDAGEYR", "DMDEDUC2", "DMDMARTL", "DMDHHSIZ", "INDFMPIR", "SEQN"]]
```

```python
df_demo.isnull().sum()
```

```python
df_demo.loc[:, "DMDEDUC2"] = df_demo["DMDEDUC2"].fillna(df_demo["DMDEDUC2"].mode()[0])
df_demo.loc[:, "DMDMARTL"] = df_demo["DMDMARTL"].fillna(df_demo["DMDMARTL"].mode()[0])
df_demo.loc[:, "INDFMPIR"] = df_demo["INDFMPIR"].fillna(df_demo["INDFMPIR"].median())
```

```python
exam = pd.read_csv("../data/raw/examination.csv")
```

```python
exam_selected = exam[["SEQN", "BMXBMI", "BPXSY1", "BPXDI1"]].copy()
```

```python
exam_selected.isnull().sum()
```

```python
exam_selected.loc[:, "BMXBMI"] = exam_selected["BMXBMI"].fillna(exam_selected["BMXBMI"].median())
exam_selected.loc[:, "BPXSY1"] = exam_selected["BPXSY1"].fillna(exam_selected["BPXSY1"].median())
exam_selected.loc[:, "BPXDI1"] = exam_selected["BPXDI1"].fillna(exam_selected["BPXDI1"].median())
```

```python
labs = pd.read_csv("../data/raw/labs.csv")
```

```python
labs_selected = labs[["SEQN", "LBDLDL", "LBXTC"]].copy()
```

```python
labs_selected.isnull().sum()
```

```python
labs_selected.loc[:, "LBDLDL"] = labs_selected["LBDLDL"].fillna(labs_selected["LBDLDL"].median())
labs_selected.loc[:, "LBXTC"] = labs_selected["LBXTC"].fillna(labs_selected["LBXTC"].median())
```

```python
merged_base = df_demo.merge(exam_selected, on = "SEQN", how = "inner")
merged_base = merged_base.merge(labs_selected, on = "SEQN", how = "inner")
```
