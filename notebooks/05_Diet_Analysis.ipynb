```python
features_diet["RIAGENDR"] = features_diet["RIAGENDR"].map({1:0, 2:1})
```

```python
feature_diet = pd.get_dummies(
    features_diet,
    columns = ["RIDRETH3", "DMDMARTL"],
    drop_first = True
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features_diet = scaler.fit_transform(features_diet)

scaled_df4 = pd.DataFrame(scaled_features_diet, columns = features_diet.columns)
scaled_df4.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia4 = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(scaled_df4)
    inertia4.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inertia4, marker = 'o')
plt.title("Elbow Method - Full Patient Dataset (Demographic + Examination + Labs + Diet)")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia / WCSS")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

for k in range(2,7):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    labels = kmeans.fit_predict(scaled_df4)
    score = silhouette_score(scaled_df4, labels)
    print(f"k={k}, silhouette score={score}")
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.mixture import GaussianMixture

X4 = scaled_df4

kmeans = KMeans(n_clusters = 2, random_state = 42)
kmeans_labels = kmeans.fit_predict(X4)
kmeans_sil = silhouette_score(X4, kmeans_labels)
print("KMeans Silhouette Score: ", kmeans_sil)

agg = AgglomerativeClustering(n_clusters = 2)
agg_labels = agg.fit_predict(X4)
agg_sil = silhouette_score(X4, agg_labels)
print("Agglomerative Silhouette Score: ", agg_sil)

gmm = GaussianMixture(n_components = 2, random_state = 42)
gmm_labels = gmm.fit_predict(X4)
gmm_sil = silhouette_score(X4, gmm_labels)
print("Gaussian Silhouette Score: ", gmm_sil)
```

```python
scaled_df4["Cluster"] = kmeans_labels
scaled_df4["Cluster"].value_counts()
```

```python
cluster_profile4 = scaled_df4.groupby("Cluster").mean()
cluster_profile4
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components = 2)
pca_data4 = pca.fit_transform(scaled_df4.drop(columns = ["Cluster"]))
pca_df4 = pd.DataFrame(pca_data4, columns = ["PC1", "PC2"])
pca_df4["Cluster"] = scaled_df4["Cluster"]

plt.figure(figsize=(8,6))
plt.scatter(pca_df4[pca_df4.Cluster == 0]["PC1"], pca_df4[pca_df4.Cluster == 0]["PC2"], label = "Cluster 0", alpha = 0.5)
plt.scatter(pca_df4[pca_df4.Cluster == 1]["PC1"], pca_df4[pca_df4.Cluster == 1]["PC2"], label = "Cluster 1", alpha = 0.5)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Visualization - Full Patient Segmentation")
plt.legend()
plt.show()
```
