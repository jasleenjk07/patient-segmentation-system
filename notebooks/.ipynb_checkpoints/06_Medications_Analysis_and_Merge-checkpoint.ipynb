```python
import pandas as pd
```

```python
base = pd.read_csv("../data/processed/base_plus_diet.csv")
```

```python
base.shape
```

```python
base.head()
```

```python
med = pd.read_csv("../data/raw/medications.csv", encoding = "latin1")
```

```python
med.shape
```

```python
med.columns
```

```python
med_summary = med.groupby("SEQN").agg(
    num_meds = ("RXDCOUNT", "max"),
    any_med_use = ("RXDUSE", lambda x: 1 if (x == 1).any() else 0)
).reset_index()
```

```python
med_summary.head()
```

```python
med_summary.shape
```

```python
med_summary["num_meds"] = med_summary["num_meds"].fillna(0)
```

```python
med_summary.isnull().sum()
```

```python
merged_med = base.merge(med_summary, on = "SEQN", how = "inner")
```

```python
merged_med.head()
```

```python
merged_med.shape
```

```python
merged_med.isnull().sum()
```

```python
merged_med[["num_meds", "any_med_use"]].describe()
```

```python
merged_med[["num_meds", "any_med_use"]].head()
```

```python
merged_med["num_meds"].value_counts().head()
```

```python
merged_med.to_csv("../data/processed/base_plus_diet_medications.csv")
```
