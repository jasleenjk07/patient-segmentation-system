```python
ques = pd.read_csv("../data/raw/questionnaire.csv")
```

```python
ques_filtered = ques.dropna(thresh = 100, axis = 1)
```

```python
ques_filtered.shape
```

```python
important_cols = [col for col in ques_filtered.columns if col.startswith(("SMQ", "ALQ", "DIQ", "MCQ", "BPQ", "PAQ"))]
```

```python
len(important_cols), important_cols[:30]
```

```python
selected_ques_cols = [
    "SEQN",
    "ALQ101",
    "ALQ130",
    "BPQ020",
    "BPQ030",
    "DIQ010",
    "DIQ160",
    "PAQ650",
    "SMQ020"
]
```

```python
ques_selected = ques_filtered[selected_ques_cols].copy()
```

```python
ques_selected.head()
```

```python
binary_cols = ["ALQ101", "BPQ020", "BPQ030", "DIQ010", "PAQ650", "SMQ020"]
ques_selected[binary_cols] = ques_selected[binary_cols].fillna(ques_selected[binary_cols].mode().iloc[0])
```

```python
ques_selected["ALQ130"] = ques_selected["ALQ130"].fillna(ques_selected["ALQ130"].median())
ques_selected["DIQ160"] = ques_selected["DIQ160"].fillna(ques_selected["DIQ160"].median())
```

```python
ques_selected.head()
```

```python
ques_selected.isnull().sum()
```

```python
merged_all = base.merge(ques_selected, on = "SEQN", how = "inner")
```

```python
merged_all.shape
```

```python
merged_all.head()
```

```python
merged_all.to_csv("../data/processed/full_patient_dataset.csv", index = False)
```
