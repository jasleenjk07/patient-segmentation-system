```python
df_demo = df[["RIDRETH3", "RIAGENDR", "RIDAGEYR", "DMDEDUC2", "DMDMARTL", "DMDHHSIZ", "INDFMPIR", "SEQN"]]
```

```python
df_demo.head()
df_demo.isnull().sum()
```

```python
df_demo.loc[:, "DMDEDUC2"] = df_demo["DMDEDUC2"].fillna(df_demo["DMDEDUC2"].mode()[0])
df_demo.loc[:, "DMDMARTL"] = df_demo["DMDMARTL"].fillna(df_demo["DMDMARTL"].mode()[0])
df_demo.loc[:, "INDFMPIR"] = df_demo["INDFMPIR"].fillna(df_demo["INDFMPIR"].median())
```

```python
df_demo.isnull().sum()
```

```python
ids = df_demo["SEQN"]
features = df_demo.drop(columns=["SEQN"]).copy()
features.head()
```

```python
features["RIAGENDR"] = features["RIAGENDR"].map({1: 0, 2: 1})
```

```python
features = pd.get_dummies(
    features,
    columns = ["RIDRETH3", "DMDMARTL"],
    drop_first = True
)
```

```python
features.info()
```

```python
features.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

scaled_df = pd.DataFrame(scaled_features, columns = features.columns)
scaled_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(1,11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,11), inertia, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia / Within cluster sum of squares')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2,8):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    labels = kmeans.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, labels)
    print(f"k={k}, silhouette score={score}")
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 42)
cluster_labels = kmeans.fit_predict(scaled_df)

scaled_df["Cluster"] = cluster_labels
scaled_df.head()
```

```python
scaled_df["Cluster"].value_counts()
```

```python
cluster_profile = scaled_df.groupby("Cluster").mean()
cluster_profile
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca_data = pca.fit_transform(scaled_df.drop(columns=["Cluster"]))

pca_df = pd.DataFrame(pca_data, columns=["PC1", "PC2"])
pca_df["Cluster"] = scaled_df["Cluster"]
pca_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,6))
plt.scatter(pca_df[pca_df.Cluster == 0]["PC1"], pca_df[pca_df.Cluster == 0]["PC2"], label = "Cluster 0", alpha = 0.6)
plt.scatter(pca_df[pca_df.Cluster == 1]["PC2"], pca_df[pca_df.Cluster == 1]["PC2"], label = "Cluster 1", alpha = 0.6)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Patient Segmentation Visualization using PCA")
plt.legend()
plt.show()
```

```python
X = scaled_df.drop(columns=["Cluster"])
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture

kmeans_labels = scaled_df["Cluster"]
kmeans_sil = silhouette_score(X, kmeans_labels)
print("KMeans Silhouette Score: ", kmeans_sil)

agg = AgglomerativeClustering(n_clusters = 2)
agg_labels = agg.fit_predict(X)
agg_sil = silhouette_score(X, agg_labels)
print("Agglomerative Silhouette Score: ", agg_sil)

gmm = GaussianMixture(n_components = 2, random_state = 42)
gmm_labels = gmm.fit_predict(X)
gmm_sil = silhouette_score(X, gmm_labels)
print("Gausian Silhouette Score: ", gmm_sil)
```
